```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv("car_dataset2.csv")
```

```python
df
```

```python
df = df[df.isnull().sum(axis=1) < 1]
```

```python
df
```

```python
df['price'] = df['price'].str.replace(' تومان', '')
df['Insurance'] = df['Insurance'].str.split('ماه', expand=True)[0].str.strip()
df['year'] = df['year'].astype(int)
df["mileage"] = df["mileage"].str.replace(',', '')
df["price"] = df["price"].str.replace(',', '')

import re

df['mileage'] = df['mileage'].apply(lambda x: int(re.sub(r'\D', '', x)))
df['price'] = df['price'].apply(lambda x: int(re.sub(r'\D', '', x)))

df['Insurance'] = df['Insurance'].astype(int)
```

```python
df
```

```python
from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()

#Shasi
df["Shasi"] = encoder.fit_transform(df['shasi'])
df.drop("shasi", axis=1, inplace=True)

#color
df["Color"] = encoder.fit_transform(df['color'])
df.drop("color", axis=1, inplace=True)

#Body
df["Body"]= encoder.fit_transform(df['body'])
df.drop("body", axis=1, inplace=True)


#Motor
df["Motor"] = encoder.fit_transform(df['motor'])
df.drop("motor", axis=1, inplace=True)
```

```python
import matplotlib.pyplot as plt

# Create a box plot of the 'price' column
plt.boxplot(df['price'])
plt.xlabel('Price')
plt.ylabel('Values')
plt.title('Box Plot of Price')

plt.show()
```

```python
# Calculate Q1 and Q3 of 'price' column
Q1 = df['price'].quantile(0.25)
Q3 = df['price'].quantile(0.75)

# Calculate IQR
IQR = Q3 - Q1

# Define the thresholds for high and low outliers
threshold_high = 0.1e10  # Adjust the high threshold as needed
threshold_low = 0.08e10  # Adjust the low threshold as needed

# Identify indices of high outliers
detect_high = np.where(df['price'] > Q3 + (1.5 * IQR))

# Identify indices of low outliers
detect_low = np.where(df['price'] < Q1 - (1.5 * IQR))

# Combine the indices of high and low outliers
outlier_indices = np.concatenate((detect_high[0], detect_low[0]))

# Remove rows with outliers
df = df.drop(df.index[outlier_indices])
```

```python
# Find indices of rows where 'price' column is not divisible evenly by 100
detect_pt=np.where(df['price']%100!=0)
for i in detect_pt:
  df=df.drop(df.index[i])
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler
sc = StandardScaler()
scaler = MinMaxScaler()
```

```python
df_scaled = sc.fit_transform(df.drop(columns='price').to_numpy())
df_scaled = pd.DataFrame(df_scaled, columns=['year', 'mileage', 'Insurance', 'Color', 'Body',])
```

```python
df_scaled
```

```python
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
from sklearn.linear_model import LinearRegression

# Split the data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(df_scaled, df['price'], test_size=0.2, random_state=42)

# Train the Linear Regression model
lin_reg = LinearRegression()
lin_reg.fit(X_train, y_train)

# Make predictions on the test set
predictions = lin_reg.predict(X_test)

# Evaluation metrics
mae, mse, rmse, r_squared = mean_absolute_error(y_test, predictions), mean_squared_error(y_test, predictions), np.sqrt(mean_squared_error(y_test, predictions)), r2_score(y_test, predictions)
rmse_cross_val = np.sqrt(-cross_val_score(lin_reg, df_scaled, df['price'], scoring="neg_mean_squared_error", cv=5)).mean()

# Print evaluation metrics
print("MAE:", mae)
print("-" * 30)

print("MSE:", mse)
print("-" * 30)

print("RMSE:", rmse)
print("-" * 30)

print("R2 Score:", r_squared)
print("-" * 30)
print("RMSE Cross-Validation:", rmse_cross_val)

# Store the model and evaluation metrics in a DataFrame
models = pd.DataFrame(columns=["Model", "MAE", "MSE", "RMSE", "R2 Score", "RMSE (Cross-Validation)"])
new_row = {"Model": "LinearRegression", "MAE": mae, "MSE": mse, "RMSE": rmse, "R2 Score": r_squared, "RMSE (Cross-Validation)": rmse_cross_val}
models = models.append(new_row, ignore_index=True)
```
